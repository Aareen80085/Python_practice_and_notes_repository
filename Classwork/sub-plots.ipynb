```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig, axes = plt.subplots(2,2)
x= np.array([1,2,3,4,5,6])
y= np.array([1,2,3,4,5,6])


axes[0,0].plot(x,y)
axes[0,1].scatter(x,y)
axes[1,1].pie(x)
axes[1,0].bar(x,y)
# axes[0,0] = plt.plot(x,y)
plt.show()
```

```python
fig, axes = plt.subplots(2,2)

x= np.array([1,2,3,4,5,6])
y= np.array([6,5,4,3,2,1])

axes[0,0].plot(x,y,label="Product A",label="Product B")
axes[0,0].scatter(x,y,label="Product A",label="Product B")

axes[0,1].scatter(x,y)



axes[1,0].bar(x,y,label="Product A",label="Product B")

axes[1,1].fill_between(x,y,label="Product A",label="Product B")

plt.show()

```
